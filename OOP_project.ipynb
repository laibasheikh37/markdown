```python
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = {}

    # to add scores for subjects
    def add_score(self, subject, score):
        self.scores[subject] = score

    # to calculate the average of scores
    def calculate_average(self):
        total_score = sum(self.scores.values())
        return total_score / len(self.scores)

    #to determine if the student is passing
    def is_passing(self):
        for score in self.scores.values():
            if score < 40:  # Assuming 40 is the passing grade threshold
                return False
        return True

    # Display student's performance
    def display_performance(self):
        average = self.calculate_average()
        passing_status = "Passing" if self.is_passing() else "Failing"
        print(f"Student: {self.name}")
        print(f"Average Score: {average:.2f}")
        print(f"Status: {passing_status}")
        print()

class PerformanceTracker:
    def __init__(self):
        self.students = {}

    # to add student and their scores
    def add_student(self, name, scores):
        student = Student(name)
        for subject, score in scores.items():
            student.add_score(subject, score)
        self.students[name] = student

    #to calculate the class average
    def calculate_class_average(self):
        total_average = 0
        for student in self.students.values():
            total_average += student.calculate_average()
        return total_average / len(self.students)

    #to display all students' performances
    def display_all_performance(self):
        for student in self.students.values():
            student.display_performance()

# Main program for input and output
def main():
    tracker = PerformanceTracker()

    while True:
        # Ask for student name
        name = input("Enter the student's name (or type 'exit' to stop): ")
        if name.lower() == 'exit':
            break

        # Input student scores
        scores = {}
        for subject in ["Math", "Science", "English"]:
            while True:
                try:
                    score = int(input(f"Enter {name}'s score in {subject}: "))
                    if 0 <= score <= 100:  # Ensure the score is between 0 and 100
                        scores[subject] = score
                        break
                    else:
                        print("Score must be between 0 and 100.")
                except ValueError:
                    print("Invalid input. Please enter an integer.")

        # Add student to the tracker
        tracker.add_student(name, scores)

    # Display student performance
    tracker.display_all_performance()

    # Display class average
    class_average = tracker.calculate_class_average()
    print(f"Class Average: {class_average:.2f}")

# Run the program
if __name__ == "__main__":
    main()
```
